<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

```python
df["INSTRUCTION_EXPEND_PER_STUDENT"] = df["INSTRUCTION_EXPENDITURE"]/df["ENROLL_ALL"]
```

Feature engineering justification: The above change was made in order to standardize the instruction expenditure amount. When schools have more students, there are likely more teachers, therefore resulting in greater overall instruction expenditure. The instruction enpenditure per student can reflect how many teachers there are per student as well as how much the teachers are paid, however more information would be needed in order to determine which of those factors are contributing more to the instruction expenditure.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[["INSTRUCTION_EXPEND_PER_STUDENT", "AVG_MATH_4_SCORE", "AVG_READING_8_SCORE"]].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=42)

```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
#R^2
model.score(X_test, y_test)
```

```python
#mean absolute error
np.mean(np.abs(model.predict(X_test)-y_test))
```

```python
#root mean squared error
np.mean((model.predict(X_test)-y_test)**2)**0.5
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('Math 8 Score')
plt.title("Model Behavior On Training Set")
```

The model used was a linear model with average 8th grade math scores as the response and averaged 8th grade reading score, 4th grade math score and instructor expenditure per student as the predictors. 70% of the data was used to train/fit the model, and the remaining 30% of the data was used to test the model to analyze it's performance.
